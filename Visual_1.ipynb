Plotting Data

```python
daycount = data['Day']

KE = data['Kenya']
CH = data['China']
BT = data['Botswana']
BR = data['Brazil']
CA = data['Canada']
MY = data['Malaysia']
SG = data['Singapore']
ID = data['Indonesia']
US = data['US']

plt.figure(figsize=(16,8))
plt.plot(daycount, KE, label="Kenya")
plt.plot(daycount, CH, label="China")
plt.plot(daycount, BT, label="Botswana")
plt.plot(daycount, BR, label="Brazil")
plt.plot(daycount, CA, label="Canada")
plt.plot(daycount, MY, label="Malaysia")
plt.plot(daycount, SG, label="Singapore")
plt.plot(daycount, ID, label="Indonwsia")
plt.plot(daycount, US, label="US")




plt.legend()
plt.title('Coronavirus Infections in Selected Countries')
plt.ylabel('Number of Infections')
plt.xlabel('Number of Days')

plt.savefig("plot.png")

plt.show()
```

Bar graph

```python
plt.figure(figsize=(16,8))
data['US'].value_counts().plot(kind='bar')
plt.legend()
plt.title('Coronavirus Infections in Selected Countries')
plt.ylabel('Number of Days')
plt.xlabel('Number of Infections')
plt.show()
```
